```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
%matplotlib inline


# Define a function to load PDOS data from a file
def load_pdos(filename):
  """Loads energy and PDOS data from a file.

  Args:
      filename (str): Name of the file containing PDOS data.

  Returns:
      tuple: Tuple containing energy and PDOS data.
  """
  data = np.loadtxt(filename)
  energy = data[:, 0]
  pdos = data[:, 1]
  return energy, pdos


# Define element names and corresponding file paths (modify as needed)
elements = ["W", "Se", "S", "Re"]
filenames = ["atom_" + element + "_tot.dat" for element in elements]


# Define background color and Fermi energy (adjust as needed)
background_color = '#F7F9F9'
fermi_energy = 1.2  # Adjust based on your data


# Load PDOS data for each element
energy_data = {}
pdos_data = {}
for element, filename in zip(elements, filenames):
  energy, pdos = load_pdos(filename)
  energy_data[element] = energy
  pdos_data[element] = pdos


# Define the amount to shift the graph leftward
shift_value = 0.0

# Shift the energy data for each element
energy_shifted = {}
for element, energy in energy_data.items():
  energy_shifted[element] = energy - shift_value


# Create the plot
plt.figure(figsize=(8, 4))
plt.figure(figsize=(8, 4), facecolor=background_color)
plt.title("Re-doped-3x3x1-System", fontsize=14)

# Plot PDOS for each element with distinct colors and labels
colors = ['#000000', '#006699', '#99ff99', '#ff9999']
for element, pdos in pdos_data.items():
  plt.plot(energy_shifted[element], pdos, linewidth=1.25, color=colors.pop(), label=element)

# Customize plot appearance
plt.yticks([])
plt.yticks([0, 5, 10, 15, 20, 25], ["0", "5", "10", "15", "20", "25"])
plt.xlabel('Energy (eV)')
plt.ylabel('PDOS(EV$^{-1}$)')
plt.axvline(x=0, linewidth=0.5, color='k', linestyle=(0, (8, 15)))

# Adjust x-axis limits based on data and shift
min_energy = min(energy.min() for energy in energy_data.values())
max_energy = max(energy.max() for energy in energy_data.values())
plt.xlim(min_energy - shift_value, max_energy - shift_value)  # Adjust based on your data range
plt.xlim(-4, 3)
plt.ylim(0, 25)
# Shade the area below Fermi level (if defined)
if fermi_energy:
  for element, pdos in pdos_data.items():
    plt.fill_between(energy_shifted[element], 0, pdos, where=(energy_shifted[element] < fermi_energy), facecolor=colors.pop(), alpha=0.25)
  plt.axvline(x=fermi_energy, linewidth=0.5, color='r', linestyle='--', label='Fermi Energy')  # Mark Fermi level

# Add legend and save the plot
plt.legend(frameon=False)
plt.savefig('Re_WSSe_pro.png', dpi=300, transparent=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd
from pycube import Cube

# Replace "path/to/your/file.cube" with the actual path to your file
cube = Cube("spincharge-density.cube")

# Access data from the cube object
grid = cube.data


def read_cube_file(filename):
    """
    Reads a .cube file generated by Quantum Espresso and extracts grid information and charge density values.

    Args:
        filename (str): Path to the .cube file.

    Returns:
        tuple: A tuple containing four NumPy arrays:
            - x: Array of x-coordinates for the grid points.
            - y: Array of y-coordinates for the grid points.
            - z: Array of z-coordinates for the grid points.
            - charge_density: Array of charge density values at each grid point, or None if an error occurs.
    """

    with open(filename, 'r') as f:
        try:
            # Skip header lines (assuming first 3 lines are comments)
            for _ in range(3):
                f.readline()

            # Read entire file content using pandas (handle mixed delimiters)
            data = pd.read_csv(f, skipinitialspace=True, sep='\s+|\t+', header=None, engine='python')

            # Extract grid dimensions (assuming first row after comments)
            try:
                nx, ny, nz = data.iloc[0]
            except (IndexError, ValueError):
                raise ValueError("Invalid data format in the file (grid dimensions).")

            # Extract charge density values (assuming all remaining data)
            charge_density = data.iloc[1:].values.flatten()  # Flatten 2D array into 1D

        except (FileNotFoundError, pd.errors.ParserError):
            print("Error reading data: The file might be corrupt or have an invalid format.")
            return None, None, None, None  # Return None for all variables if an error occurs

    # Convert grid dimensions to integers (assuming numerical format)
    nx = int(nx)
    ny = int(ny)
    nz = int(nz)

    # Reshape charge density into 3D array based on grid dimensions
    charge_density = np.reshape(charge_density, (nz, ny, nx))

    return x, y, z, charge_density


# Specify the path to your .cube file (replace with your actual filename)
filename = "spincharge-density.cube"


# Read data from the cube file
try:
    x, y, z, charge_density = read_cube_file(filename)
except ValueError as e:
    print("Error reading data:", e)
    exit()  # Exit the program if data reading fails

# Choose visualization approach: (A) Volumetric or (B) Isosurface
print("Choose visualization approach:")
print(" (A) Volumetric Model (entire charge density distribution)")
print(" (B) Isosurface Model (specific charge density region)")
choice = input("Enter your choice (A or B): ").upper()

# Check if data reading was successful before proceeding
if charge_density is None:
    print("Error: Data reading failed. Visualization cannot be performed.")
    exit()  # Exit the program if data reading fails

def visualize_charge_density(charge_density, choice, isosurface_value=None):
    """
    Visualizes the charge density distribution based on the chosen approach.

    Args:
        charge_density (np.ndarray): 3D array of charge density values.
        choice (str): Visualization approach (A or B).
        isosurface_value (float, optional): Isosurface value for choice (B). Defaults to None.
    """

    if choice == "A":
        # Volumetric Model
        fig = plt.figure(figsize=(10, 6))
        ax = fig.add_subplot(111, projection='3d')

        # Separate negative and positive densities (optional)
        negative_density = charge_density.clip(upper=0)
        positive_density = charge_density.clip(lower=0)

        # Visualize densities (adjust transparency as needed)
        ax.voxellimage(negative_density, cmap='coolwarm', alpha=0.5, label='Negative Charge Density')
        ax.voxellimage(positive_density, cmap='coolwarm', alpha=0.5, label='Positive Charge Density')

        # Customize plot
        ax.set_xlabel("X (a.u.)")
        ax.set_ylabel("Y (a.u.)")
        ax.set_zlabel("Z (a.u.)")
        ax.set_title("Charge Density Distribution (Volumetric)")
        ax.legend()  # Add legend for negative and positive densities
        plt.show()

    # Add the missing indentation level here (one level less than 'if' block)
    elif choice == "B":
        # Isosurface Model
        fig, ax = plt.subplots(figsize=(10, 6))  # Create figure and 3D axes

        # Use trimesh (triangle mesh) for isosurface generation
        from skimage.measure import marching_cubes

        verts, faces = marching_cubes(charge_density, isosurface_value)  # Generate mesh for isosurface

        # Plot the isosurface
        ax.plot_trisurf(verts[:, 0], verts[:, 1], verts[:, 2], triangles=faces, cmap='viridis', linewidth=0.1)  # Adjust colormap and line width as needed

        # Customize plot
        ax.set_xlabel("X (a.u.)")
        ax.set_ylabel("Y (a.u.)")
        ax.set_zlabel("Z (a.u.)")
        ax.set_title("Charge Density Distribution (Isosurface)")
        plt.show()
```
