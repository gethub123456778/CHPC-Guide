```python
import re

def parse_scf_output(filename):
  """
  Parses the SCF.OUT file and extracts relevant energies (if found).

  Args:
      filename (str): Path to the SCF.OUT file.

  Returns:
      dict: Dictionary containing extracted energies (eV) or None if parsing failed.
  """
  try:
    with open(filename, "r") as f:
      output = f.read()

    # Compile regular expressions for efficiency
    bandgap_regex = r"Bandgap \(eV\):\s*(\d+\.\d+)"
    fermi_regex = r"Fermi energy\s*=\s*(\d+\.\d+)"
    homo_regex = r"highest occupied eigenvalue\s*=\s*(\d+\.\d+)"
    lumo_regex = r"lowest unoccupied eigenvalue\s*=\s*(\d+\.\d+)"

    bandgap_match = re.search(bandgap_regex, output)
    fermi_match = re.search(fermi_regex, output)
    homo_match = re.search(homo_regex, output)
    lumo_match = re.search(lumo_regex, output)

    # Check if all matches were found
    if all((bandgap_match, fermi_match, homo_match, lumo_match)):
      bandgap = float(bandgap_match.group(1))
      fermi_energy = float(fermi_match.group(1))
      homo_energy = float(homo_match.group(1))
      lumo_energy = float(lumo_match.group(1))

      electron_affinity = fermi_energy - homo_energy
      ionization_energy = lumo_energy - fermi_energy

      return {
        "bandgap": bandgap,
        "fermi_energy": fermi_energy,
        "electron_affinity": electron_affinity,
        "ionization_energy": ionization_energy,
      }
    else:
      print("Error: Could not find all required keywords in SCF.OUT file.")
      return None

  except (FileNotFoundError, AttributeError):
    print("Error: Failed to open or read SCF.OUT file.")
    return None

if __name__ == "__main__":
  # Replace "SCF.OUT" with your actual filename
  results = parse_scf_output("scf.out")

  if results:
    bandgap = results["bandgap"]
    fermi_energy = results["fermi_energy"]
    electron_affinity = results["electron_affinity"]
    ionization_energy = results["ionization_energy"]

    print("Bandgap:", bandgap, "eV")
    print("Vacuum level:", fermi_energy, "eV")
    print("Electron affinity:", electron_affinity, "eV")
    print("Ionization energy:", ionization_energy, "eV")
  else:
    print("Error: Parsing failed.")
```
